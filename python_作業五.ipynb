```python
import os
import google_auth_oauthlib.flow
import googleapiclient.discovery
import googleapiclient.errors
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import numpy as np

# Set your API key
os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1"
os.environ["OAUTHLIB_RELAX_TOKEN_SCOPE"] = "1"
api_key = 'AIzaSyCeGckH2g0VjnPTdcStMTUBlis3TUlLrnE'

# Initialize the YouTube Data API client
youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=api_key)

# Video ID for fetching comments
video_id = '__EwIUp9lq4'

# Define the keywords to check in the comments
positive_keyword = '明確'
useful_keyword = '實用'
creative_keyword = '創意'
clear_keyword = '清楚'
complete_keyword = '完整'

# Fetch comments sentiment analysis based on keywords
positive_comments = 0
useful_comments = 0
creative_comments = 0
clear_comments = 0
complete_comments = 0

next_page_token = None

while True:
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        textFormat="plainText",
        pageToken=next_page_token
    )
    response = request.execute()

    for comment in response["items"]:
        text_display = comment["snippet"]["topLevelComment"]["snippet"]["textDisplay"].lower()

        # Calculate sentiment scores based on keywords
        positive_comments += text_display.count(positive_keyword)
        useful_comments += text_display.count(useful_keyword)
        creative_comments += text_display.count(creative_keyword)
        clear_comments += text_display.count(clear_keyword)
        complete_comments += text_display.count(complete_keyword)

    next_page_token = response.get("nextPageToken")

    if next_page_token is None:
        break

# Display sentiment statistics
print("Comments with '{}' keyword:", positive_comments)
print("Comments with '{}' keyword:", useful_comments)
print("Comments with '{}' keyword:", creative_comments)
print("Comments with '{}' keyword:", clear_comments)
print("Comments with '{}' keyword:", complete_comments)

# Set font
font_path = "NotoSansTC-Regular.ttf"  # 字體文件的路徑
prop = fm.FontProperties(fname=font_path)

# Convert sentiment statistics to radar chart
attributes = ['Comments with "{}"'.format(positive_keyword),
              'Comments with "{}"'.format(useful_keyword),
              'Comments with "{}"'.format(creative_keyword),
              'Comments with "{}"'.format(clear_keyword),
              'Comments with "{}"'.format(complete_keyword)]
scores = [positive_comments, useful_comments, creative_comments, clear_comments, complete_comments]

# Calculate angles
num_attributes = len(attributes)
angles = np.linspace(0, 2 * np.pi, num_attributes, endpoint=False).tolist()
angles += angles[:1]

# Ensure scores have the same number of elements as angles
scores += scores[:1]

# Set up the plot
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Plot the radar chart
ax.fill(angles, scores, color='darkred', alpha=0.25)
ax.plot(angles, scores, color='darkred', linewidth=2, linestyle='solid')
ax.fill(angles, scores, color='darkred', alpha=0.2)

# Add labels with custom font
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(attributes, fontproperties=prop)

# Display the radar chart
plt.show()
```
