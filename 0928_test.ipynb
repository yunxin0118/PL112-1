```python
import pandas as pd
data = pd.read_csv('adult.data.csv', header=None)
data.head()
```

```python
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
              'marital_status', 'occupation', 'relationship', 'race', 'sex',
              'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
data.head()
```

# 每周工作時間

```python
hour_income = data.groupby('income')['hours_per_week'].agg(['mean', 'std', 'median'])
hour_income
```

每周工作時間與收入是否有關係?

```python
income_hour = data[['income', 'hours_per_week']]

import plotly.express as px

fig = px.violin(income_hour, y="hours_per_week", x="income", color="income",
                box=True, points="all",
          hover_data=income_hour.columns)
fig.show()
```

# 教育程度

```python
income_by_job = data.groupby(['education', 'income'])['sex'].count().unstack()
income_by_job
```

哪些教育程度更容易賺到高薪?

```python
income_by_job['high_income_ratio'] = income_by_job[' >50K'] / income_by_job.sum(axis=1)
income_by_job['low_income_ratio'] = income_by_job[' <=50K'] / income_by_job.sum(axis=1)

income_by_job
```

教育程度是否影響收入?
